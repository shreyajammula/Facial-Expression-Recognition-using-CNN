### Import Necessary Libraries

```python
import numpy as np
from tensorflow.keras.models import model_from_json
from tensorflow.keras.models import load_model
import h5py
from PIL import Image
```

ignore the warining if any while importing

### Load saved model JSON architecture

```python
json_file = open('face_cnn_model_architecture.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
```

### Load the saved model weights

```python
loaded_model.load_weights('face_cnn_model_weights.h5')
```

### Compile the model (required before making predictions)

```python
loaded_model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

### Define the emotion labels

```python
emo = ['Angry', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
```

### Define Function to Pre-Process and Predict Image

```python
def predict_emotion_from_image_loaded_model(image_path, model):
    # Read image using PIL
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    img = img.resize((48, 48))

    # Convert PIL image to numpy array
    img_array = np.array(img)

    # Expand dimensions to match the model input shape
    img_array = np.expand_dims(img_array, axis=0)
    img_array = np.expand_dims(img_array, axis=-1)

    # Normalize pixel values
    img_array = img_array / 255.0

    emotion_idx = np.argmax(model.predict(img_array))
    predicted_emotion = emo[emotion_idx]
    
    return predicted_emotion
```

### Test the loaded model

```python
image_path = input('Enter your image file path: ')  # Replace with the actual image path
predicted_emotion = predict_emotion_from_image_loaded_model(image_path, loaded_model)
print(f'Predicted Emotion: {predicted_emotion}')
```
